```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image_path = '/content/drive/MyDrive/Colab Notebooks/Data Area Mangrove/area-1-mangrove_sig.png'  # Ganti dengan path ke gambar Anda
image = cv2.imread(image_path)

# Verify that the image has been loaded
if image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply median filtering to remove noise
    median_filtered = cv2.medianBlur(gray, 5)

    # Apply adaptive thresholding
    adaptive_thresh = cv2.adaptiveThreshold(
        median_filtered, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)

    # Apply morphological opening to separate objects
    kernel = np.ones((3, 3), np.uint8)
    morph_open = cv2.morphologyEx(adaptive_thresh, cv2.MORPH_OPEN, kernel)

    # Find contours
    contours, hierarchy = cv2.findContours(
        morph_open, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the original image
    image_contours = image.copy()
    cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)

    # Count the number of contours (potential mangrove trees)
    num_mangrove_trees = len(contours)

    # Display the results
    print(f'Number of mangrove trees detected: {num_mangrove_trees}')

    # Plotting the images
    plt.figure(figsize=(15, 8))

    plt.subplot(2, 3, 1)
    plt.title('Original Image')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    plt.subplot(2, 3, 2)
    plt.title('Grayscale Image')
    plt.imshow(gray, cmap='gray')

    plt.subplot(2, 3, 3)
    plt.title('Median Filtered Image')
    plt.imshow(median_filtered, cmap='gray')

    plt.subplot(2, 3, 4)
    plt.title('Adaptive Thresholding')
    plt.imshow(adaptive_thresh, cmap='gray')

    plt.subplot(2, 3, 5)
    plt.title('Morphological Opening')
    plt.imshow(morph_open, cmap='gray')

    plt.subplot(2, 3, 6)
    plt.title('Contours')
    plt.imshow(cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB))

    plt.tight_layout()
    plt.show()
```
